<span style="color:red">TODO: Create a function to fix the digitation errors for "DIA" column.</span>

```python
dia_semana =  ['Lunes','Martes','Miércoles','Jueves','Viernes','Sábado','Domingo']
def fix_day_of_week(string):
  Dict = {}
  for x in dia_semana:
    Dict[x] = pylev.levenshtein(x, string)
  return min(Dict, key=Dict.get)
```

```python
homicides_df["DIA"] = homicides_df["DIA"].apply(fix_day_of_week)
homicides_df
```

<span style="color:red">TODO: Fix the "AÑO DE NACIMIENTO" column using the column "EDAD".</span>

```python
def fix_born_year(born_year,age):
  year_born = 2022 - age
  if (year_born+1 == born_year or year_born-1 == born_year ):
    return born_year
  return year_born
  

homicides_df["AÑO DE NACIMIENTO"] = homicides_df.apply(lambda x: fix_born_year(x["AÑO DE NACIMIENTO"], x["EDAD"]), axis=1)
```
